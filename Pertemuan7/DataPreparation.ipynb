### 5. Membersihkan Dataset dengan ApplyMap

```python
university_town = []
with open('university_towns.txt') as file:
    for line in file:
        if '[edit]' in line:
            state = line
        else :
            university_town.append((state,line))

university_town[:5]
```

```python
dfCity = pd.DataFrame(university_town, columns=['State', 'Region Name'])
dfCity.head()
```

```python
def getCity(item):
    if '(' in item:
        return item[:item.find('(')]
    elif '[' in item:
        return item[:item.find('[')]
    else:
        return item

```

```python
dfCity =  dfCity.applymap(getCity)
```
