### Pie Chart

```python
import matplotlib.pyplot as plt

tasteVariant = ('Chocolate', 'Strawberry', 'Vanilla', 'Taro', 'Red Velvet')
dataDummy = (12, 15, 3, 11, 9)

plt.pie(dataDummy, labels = tasteVariant)
plt.show
```

```python
plt.pie(dataDummy, labels = tasteVariant, autopct='%1.1f%%')
plt.show
```

```python
color = ('#733d00','#E32E34','#FFFFFF', '#ffc1ff', 	'#9c0000')
plt.pie(dataDummy, labels = tasteVariant, autopct='%1.1f%%', colors=color)
plt.show
```

```python
explode = (0,0,0.15,0,0)

plt.pie(dataDummy, labels = tasteVariant, autopct='%1.1f%%', colors=color, explode=explode)
plt.show
```

```python
plt.pie(dataDummy, labels = tasteVariant, autopct='%1.1f%%', colors=color, explode=explode, shadow=True)
plt.show
```

```python

plt.pie(dataDummy, labels = tasteVariant, autopct='%1.1f%%', colors=color, explode=explode, shadow=True)
plt.title('Ice Cream dengan Penjualan Terendah')
plt.show
```

### BarChart

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

country = ('Argentina', 'Brazil','Chile', 'Denmark','England','Finland', 'Germany','Hongaria', 'Indonesia','Jepang')
population = (45380000, 212600000, 19120000, 5831000, 55980000, 5531000, 83240000, 9750000, 273500000, 125000000)
x_coor = np.arange(len(country))

plt.bar(x_coor, population)
plt.show
```

```python
plt.bar(x_coor, population, tick_label=country)
plt.xticks(rotation=90)
plt.show
```

```python
plt.bar(x_coor, population, tick_label=country)
plt.xticks(rotation=90)
plt.ylabel('Population (Million)')
plt.show
```

```python
plt.title('Top 10 Country Population')
plt.bar(x_coor, population, tick_label=country)
plt.xticks(rotation=90)
plt.ylabel('Population (Million)')
plt.show
```

```python

country = ('Argentina', 'Brazil','Chile', 'Denmark','England','Finland', 'Germany','Hongaria', 'Indonesia','Jepang')
population = (45380000, 212600000, 19120000, 5831000, 55980000, 5531000, 83240000, 9750000, 273500000, 125000000)

df = pd.DataFrame({'Country':country, 'Population':population})
df.sort_values(by='Population', inplace=True)

x_coor = np.arange(len(df))

plt.title('Top 10 Country Population')
plt.bar(x_coor, df['Population'], tick_label=df['Country'])
plt.xticks(rotation=90)
plt.ylabel('Population (Million)')
plt.show
```

### Challenges 2
Ubah sort dari ASC menjadi DSC

```python
country = ('Argentina', 'Brazil','Chile', 'Denmark','England','Finland', 'Germany','Hongaria', 'Indonesia','Jepang')
population = (45380000, 212600000, 19120000, 5831000, 55980000, 5531000, 83240000, 9750000, 273500000, 125000000)

df = pd.DataFrame({'Country':country, 'Population':population})
df.sort_values(by='Population', inplace=True, ascending=False)

x_coor = np.arange(len(df))

plt.title('Top 10 Country Population')
plt.bar(x_coor, df['Population'], tick_label=df['Country'])
plt.xticks(rotation=90)
plt.ylabel('Population (Million)')
plt.show
```

Memberi Warna pada data ketiga dari terakhir

```python
country = ('Argentina', 'Brazil','Chile', 'Denmark','England','Finland', 'Germany','Hongaria', 'Indonesia','Jepang')
population = (45380000, 212600000, 19120000, 5831000, 55980000, 5531000, 83240000, 9750000, 273500000, 125000000)

df = pd.DataFrame({'Country':country, 'Population':population})
df.sort_values(by='Population', inplace=True)

x_coor = np.arange(len(df))
colors = ['#0000FF' for _ in range(len(df))]
colors[-3] = '#FF0000'

plt.title('Top 10 Country Population')
plt.bar(x_coor, df['Population'], tick_label=df['Country'], color=colors)
plt.xticks(rotation=90)
plt.ylabel('Population (Million)')
plt.show
```

Tambah Figure

```python
country = ('Argentina', 'Brazil','Chile', 'Denmark','England','Finland', 'Germany','Hongaria', 'Indonesia','Jepang')
population = (45380000, 212600000, 19120000, 5831000, 55980000, 5531000, 83240000, 9750000, 273500000, 125000000)

df = pd.DataFrame({'Country':country, 'Population':population})
df.sort_values(by='Population', inplace=True)

x_coor = np.arange(len(df))
colors = ['#0000FF' for _ in range(len(df))]
colors[-3] = '#FF0000'

plt.figure(figsize=(20,10))

plt.title('Top 10 Country Population')
plt.bar(x_coor, df['Population'], tick_label=df['Country'], color=colors)
plt.xticks(rotation=90)
plt.ylabel('Population (Million)')
plt.show
```
