#### 2. Hierarchical Clustering

Sumber Data : https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/cars_clus.csv

```python
nameFile = "cars_clus.csv"

#Baca Dataframe

dataHC = pd.read_csv(nameFile)

# Banyak Baris dan Kolom
print("Banyak (baris, kolom): ", dataHC.shape)
dataHC.head()
```

Cleaning

```python
print("Banyaknya data sebelum di bersihkan: ", dataHC.size)
dataHC[['sales','resale','type','price','engine_s','horsepow','wheelbas','width','length','curb_wgt','fuel_cap','mpg','lnsales']] = dataHC[['sales','resale','type','price','engine_s','horsepow','wheelbas','width','length','curb_wgt','fuel_cap','mpg','lnsales']].apply(pd.to_numeric, errors='coerce')
dataHC = dataHC.dropna()
dataHC = dataHC.reset_index(drop=True)
print("Banyaknya data sebelum di bersihkan: ", dataHC.size)
```

Kelompokkan Fiture 

```python
fiture = dataHC[['engine_s','width','length','fuel_cap','mpg']]
# fiture.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
x = fiture.values #returns a numpy array
min_max_scaler = MinMaxScaler()
fiture_minmax = min_max_scaler.fit_transform(x)
fiture_minmax[:5]
```

Import SCIPY

```python
import scipy
leng = fiture_minmax.shape[0]
D = np.zeros([leng,leng])

for i in range(leng):
    for j in range(leng):
        D[i,j] = scipy.spatial.distance.euclidean(fiture_minmax[i],fiture_minmax[j])
```

```python
#Import Python Library
import pylab
import scipy.cluster.hierarchy as sch
z = sch.linkage(D, method = 'complete')
```

```python
from scipy.cluster.hierarchy import fcluster
max_d = 3
clusters = fcluster(z, max_d, criterion='distance')
clusters
```

```python
# Kembali Menggunakan FCluster
k = 3
clusters_new = fcluster(z, k, criterion='maxclust')
clusters_new
```

```python
# Lihat Gambar 
gambar = pylab.figure(figsize=(20,70))
def llf(id):
    return '[%s %s %s]' % (dataHC['manufact'][id], dataHC['model'][id], int(float(dataHC['type'][id])))
gambar_dendro = sch.dendrogram(z, leaf_label_func=llf, leaf_rotation=0, leaf_font_size =16, labels = dataHC['model'], orientation = 'right')
```

#### 3. DBSCAN

```python
# Import Library

from sklearn.cluster import DBSCAN
```

```python
def createDataPoints(centroid, numSamples, clusterDeviation):
    X,y = make_blobs(n_samples=numSamples, centers=centroid, cluster_std=clusterDeviation)
    X = StandardScaler().fit_transform(X)
    return X , y
```

```python
X, y = createDataPoints([[4,3], [2,-1], [-1,4], [12,4]], 1500, 0.5)
```

#### Pemodelan
DBSCAN kepanjangannya Density-Based Spatial Clustering of Applications with Noise. Nah ini salah satu teknik clustering yang bekerja berdasarkan tingkat <b>density of object</b>. Intinya semua point/titik dalam sebuah cluster, harus saling berdekatan dengan point/titik lainnya.
<p>Cara kerjanya berdasarkan 2 parameter: <b>Epsilon</b> (jarak antar cluster atau radius) dan <b>Minimum Points</b>

```python
epsilon = 0.3
minimumSamples = 7
db = DBSCAN(eps=epsilon, min_samples=minimumSamples).fit(X)
labels = db.labels_
labels
```

```python
# Buat Array
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
core_samples_mask
```

```python
# cek banyak cluster dalam label, jika ada 'noise' maka abaikan saja

n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_clusters_
```

```python
# Buat jadi unique (tidak ada data yang sama / berulang)
unique_labels = set(labels)
unique_labels
```

Visualisasi Data 

```python
# Setting Warna
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
colors
```

```python
# Buat tiap titik / points menjadi berwarna 
for k, col in zip(unique_labels, colors):
    if(k == -1):
        # Setiap noise diberi warna hitam
        col = 'k'
        
class_member_mask = (labels == k)

# Gambar tiap titik yang tercluster
xy = X[class_member_mask & core_samples_mask]
plt.scatter(xy[:, 0], xy[:, 1],s=50,c=col, marker=u'o', alpha=0.5)

#Gambar titik yang diluar cluster
xy = X[class_member_mask & ~core_samples_mask]
plt.scatter(xy[:, 0], xy[:, 1],s=50,c=col, marker=u'o', alpha=0.5)
```

```python
from sklearn.cluster import KMeans 
k = 3
k_means3 = KMeans(init = "k-means++", n_clusters = k, n_init = 12)
k_means3.fit(X)
fig = plt.figure(figsize=(6, 4))
ax = fig.add_subplot(1, 1, 1)
for k, col in zip(range(k), colors):
    my_members = (k_means3.labels_ == k)
    plt.scatter(X[my_members, 0], X[my_members, 1],  c=col, marker=u'o', alpha=0.5)
plt.show()
```
