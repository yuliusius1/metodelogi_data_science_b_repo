### 2. Pie Chart
Alasan Memilih Pie Chart
> Alasan saya memilih pie chart karena data yang diolah tidak banyak yaitu hanya 5 data saja, lalu untuk pencarian persentasi kemunculan, akan lebih mudah menggunakan pie chart karena dapat menggunakan autopct='%1.1f%%'. Selain itu kita juga dapat melakukan explode terhadap data yang dipilih (kopiko)

Peluang Kemunculan Permen Kopiko
> Dapat dilihat pada chart, persentase peluang kemunculan permen Kopiko adalah sebesar 15% atau 3/20 

```python
nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
jumlah_permen = [52, 39, 78, 13, 78]
explode = (0,0.1,0,0,0)
color = ('#733d00','#E32E34','#FFFFFF', '#ffc1ff', 	'#9c0000')

plt.title("Pie Chart Kantong Permen")
plt.pie(jumlah_permen, labels=nama_permen, explode=explode,  autopct='%1.1f%%', colors=color)
plt.show()
```

### 3. Bar Chart

Alasan memilih Bar Chart 
> Saya memilih bar chart karena studi kasus no 3 menghitung 3 makanan dengan jumlah penjualan terendah dan diminta untuk menunjukan visualisasinya, maka dari itu chart yang tepat untuk digunakan dalam menyelesaikan soal ini adalah Bar Chart

Buat visualisasi yang menunjukkan popularitas relatif dari makanan penutup.
> Dapat dilihat dibawah bahwa Bar chart saya urutkan dari makanan dengan popularitas tertinggi (Ice Cream) ke popularitas terendah (Kue Wajik) 

Sebutkan tiga makanan penutup yang harus disingkirkan.
> Karena data berurutan dari popularitas tertinggi ke popularitas terendah, maka kita hanya melihat dari 3 terbawah saja untuk mencari 3 makanan yang harus disingkirkan. maka dari itu kita dapat menyingkirkan <b> Kue Wajik, Pastel dan Puding Vanilla </b>

```python
dataPenjualan = {'Donat': 14, 'Pastel': 5,'Kue Coklat': 12,'Ice Cream': 19,'Puding Vanila': 6,'Brownies': 8,'Puding Strawberi': 12,'Puding Coklat': 9,'Ice Cream Nutela': 10,'Kue Coklat-Keju': 17,'Kue Wajik': 2,'Kue Sus': 9,'Mochi': 13}
df_dataPenjualan = pd.DataFrame(list(dataPenjualan.items())).rename(columns={0:"Nama Makanan", 1:"Jumlah Penjualan"})
df_dataPenjualan.sort_values(by="Jumlah Penjualan", inplace=True)
x_coor = np.arange(len(df_dataPenjualan))

colors = []
for _ in range(len(df_dataPenjualan)) :
    colors.append('#FF0000' if _ < 3 else '#0000FF')

plt.barh(x_coor,df_dataPenjualan["Jumlah Penjualan"], tick_label=df_dataPenjualan["Nama Makanan"], color=colors)
plt.ylabel("Jumlah Penjualan")
plt.title("Jumlah Penjualan Makanan Kantin Biru")
plt.show()
```

### 4. HeatMap
